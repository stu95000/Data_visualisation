```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

file_path = "WHO_TB_Data_4565.xlsx"
df = pd.read_excel(file_path)

df.head()
```

```python
df.columns
```

```python
df.head(20)
```

```python
# Set the first row as column headers
df.columns = df.iloc[0]

# Remove the header row from the data
df = df[1:]

# Reset the index
df = df.reset_index(drop=True)

df.head()
```

```python
df.columns
```

```python
# Renaming the first column to a simple name
df = df.rename(columns={df.columns[0]: "Country"})

# Converting year columns from wide to long format
df_long = df.melt(
    id_vars="Country",
    var_name="Year",
    value_name="TB_per_100k"
)

# Cleaning the Year column (turn "Year 1991" into 1991)
df_long["Year"] = df_long["Year"].str.replace("Year ", "", regex=False).astype(int)

# Converting TB values to numeric (handles any weird text)
df_long["TB_per_100k"] = pd.to_numeric(df_long["TB_per_100k"], errors="coerce")

df_long.head()
```

```python
df_long.info()
```

```python
df_long.isna().sum()
```

```python
df_long.describe()
```

```python
plt.figure(figsize=(10,6))

# Global average TB rate per year
global_trend = df_long.groupby("Year")["TB_per_100k"].mean()

plt.plot(global_trend.index, global_trend.values)
plt.title("Global Average Tuberculosis Incidence (19912007)")
plt.xlabel("Year")
plt.ylabel("TB cases per 100,000 population")

plt.show()
```

```python
# Calculate average TB incidence per country across all years
country_avg = (
    df_long
    .groupby("Country")["TB_per_100k"]
    .mean()
    .sort_values(ascending=False)
)

# Top 10 highest TB burden countries
top_10_high = country_avg.head(10)

# Top 10 lowest TB burden countries
top_10_low = country_avg.tail(10)

top_10_high, top_10_low
```

```python
plt.figure(figsize=(10,6))
top_10_high.plot(kind="bar")
plt.title("Top 10 Countries with Highest Average TB Incidence (19912007)")
plt.xlabel("Country")
plt.ylabel("TB cases per 100,000 population")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(10,6))
top_10_low.plot(kind="bar")
plt.title("Top 10 Countries with Lowest Average TB Incidence (19912007)")
plt.xlabel("Country")
plt.ylabel("TB cases per 100,000 population")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Find the row with the maximum TB incidence
peak_row = df_long.loc[df_long["TB_per_100k"].idxmax()]

peak_row
```

```python
# Filter data for the peak country
peak_country = peak_row["Country"]

peak_country_data = (
    df_long[df_long["Country"] == peak_country]
    .sort_values("Year")
)

plt.figure(figsize=(10,6))
plt.plot(
    peak_country_data["Year"],
    peak_country_data["TB_per_100k"]
)
plt.title(f"TB Incidence Over Time in {peak_country}")
plt.xlabel("Year")
plt.ylabel("TB cases per 100,000 population")
plt.show()
```

```python
# Prepare Top 10 countries by average TB incidence
top10_countries = (
    df_long.groupby("Country")["TB_per_100k"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .index
)

df_top10 = df_long[df_long["Country"].isin(top10_countries)]
```

```python
from plotly.subplots import make_subplots
import plotly.express as px

# Chart 1: Global trend with time range slider
global_df = df_long.groupby("Year", as_index=False)["TB_per_100k"].mean()
fig1 = px.line(
    global_df,
    x="Year",
    y="TB_per_100k",
    labels={"TB_per_100k": "TB cases per 100,000 population"}
)
fig1.update_xaxes(rangeslider_visible=True)

# Safety: rebuild df_top10 (Top 10 countries) if needed
top10_countries = (
    df_long.groupby("Country")["TB_per_100k"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .index
)
df_top10 = df_long[df_long["Country"].isin(top10_countries)]

# Chart 2: Top 10 country trends
fig2 = px.line(
    df_top10,
    x="Year",
    y="TB_per_100k",
    color="Country",
    labels={"TB_per_100k": "TB cases per 100,000 population"}
)

# Build Top 10 highest average TB as a DataFrame (safe)
top10_high_df = (
    df_long.groupby("Country", as_index=False)["TB_per_100k"]
    .mean()
    .sort_values("TB_per_100k", ascending=False)
    .head(10)
)

# Chart 3: Top 10 highest TB burden (bar chart)
fig3 = px.bar(
    top10_high_df,
    x="Country",
    y="TB_per_100k",
    labels={"TB_per_100k": "TB cases per 100,000 population"}
)

# Combine into one dashboard
dashboard = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Global TB Trend (19912007)",
        "TB Trends: Top 10 High-Burden Countries",
        "Top 10 Countries by Average TB Incidence",
        ""
    )
)

for trace in fig1.data:
    dashboard.add_trace(trace, row=1, col=1)

for trace in fig2.data:
    dashboard.add_trace(trace, row=1, col=2)

for trace in fig3.data:
    dashboard.add_trace(trace, row=2, col=1)

dashboard.update_layout(
    title_text="Interactive Dashboard: Global Tuberculosis Incidence (19912007)",
    height=800,
    showlegend=True
)

dashboard.show()
```
